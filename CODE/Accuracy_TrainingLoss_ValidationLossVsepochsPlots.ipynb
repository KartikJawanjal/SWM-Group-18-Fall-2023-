```python
import torch
```

```python
file_path = 'amazon_daily.pt'

# Load the file
loaded_data = torch.load(file_path)

# Assuming your dictionary is stored in the loaded data
your_dictionary = loaded_data
```

```python
print(your_dictionary)
```

```python
import matplotlib.pyplot as plt



# The dictionary containing accuracy values
accuracy_data = loaded_data

# Extracting the validation accuracy values
val_accuracy = accuracy_data['val_accuracy']

# Generating x-axis values (epochs)
epochs = range(1, len(val_accuracy) + 1)

```

```python
val_accuracy = loaded_data['val_accuracy']
val_loss = loaded_data['val_loss']
train_loss = loaded_data['train_loss']
# Generating x-axis values (epochs)
epochs = range(1, len(val_accuracy) + 1)

# Plotting the metrics
plt.figure(figsize=(12, 6))

# Plotting validation accuracy
plt.subplot(1, 3, 1)
plt.plot(epochs, val_accuracy, label='Validation Accuracy', marker='o', color='blue')
plt.title('Validation Accuracy Over Epochs')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.grid(True)

# Plotting validation loss
plt.subplot(1, 3, 2)
plt.plot(epochs, val_loss, label='Validation Loss', marker='o', color='red')
plt.title('Validation Loss Over Epochs')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.grid(True)

# Plotting validation loss
plt.subplot(1, 3, 3)
plt.plot(epochs, train_loss, label='train Loss', marker='o', color='red')
plt.title('training Loss Over Epochs')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.grid(True)

print("AMAZON_DAILY")
plt.tight_layout()
plt.show()
```
