```python
import torch
```

```python
file_path = 'amazon_daily.pt'
loaded_data = torch.load(file_path)
your_dictionary = loaded_data
```

```python
print(your_dictionary)
```

```python
import matplotlib.pyplot as plt

accuracy_data = loaded_data
val_accuracy = accuracy_data['val_accuracy']
epochs = range(1, len(val_accuracy) + 1)
```

```python
val_accuracy = loaded_data['val_accuracy']
val_loss = loaded_data['val_loss']
train_loss = loaded_data['train_loss']
# Generating x-axis values (epochs)
epochs = range(1, len(val_accuracy) + 1)

# Plotting the metrics
plt.figure(figsize=(12, 6))

# Plotting validation accuracy
plt.subplot(1, 3, 1)
plt.plot(epochs, val_accuracy, label='Validation Accuracy', marker='o', color='blue')
plt.title('Validation Accuracy Over Epochs')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.grid(True)

# Plotting validation loss
plt.subplot(1, 3, 2)
plt.plot(epochs, val_loss, label='Validation Loss', marker='o', color='red')
plt.title('Validation Loss Over Epochs')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.grid(True)

# Plotting validation loss
plt.subplot(1, 3, 3)
plt.plot(epochs, train_loss, label='train Loss', marker='o', color='red')
plt.title('training Loss Over Epochs')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.grid(True)

print("AMAZON_DAILY")
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
data = {
    'Models': ['LR Apple Hourly', 'LR Apple Daily', 'LR Amazon Hourly', 'LR Amazon Daily',
               'SVM Apple Hourly', 'SVM Apple Daily', 'SVM Amazon Hourly', 'SVM Amazon Daily'],
    'Accuracy': [0.632, 0.73, 0.6525, 0.765, 0.636, 0.721, 0.647, 0.724],
}
colors = ['red', 'green', 'blue', 'orange', 'purple', 'pink', 'brown', 'gray']
df = pd.DataFrame(data)

# Plotting
plt.figure(figsize=(10, 6))
plt.bar(df['Models'], df['Accuracy'], color=colors)
plt.title('Accuracy for Different Models')
plt.xlabel('Models')
plt.ylabel('Accuracy')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

data = {
    'Models': ['LR Apple Hourly', 'LR Apple Daily', 'LR Amazon Hourly', 'LR Amazon Daily',
               'SVM Apple Hourly', 'SVM Apple Daily', 'SVM Amazon Hourly', 'SVM Amazon Daily'],
    'F1_Score_0': [0.65, 0.72, 0.6175, 0.7525, 0.626, 0.714, 0.59, 0.696],
    'F1_Score_1': [0.62, 0.74, 0.68, 0.775, 0.65, 0.73, 0.692, 0.748],
}

df = pd.DataFrame(data)

colors = ['lightcoral', 'lightgreen']

# Plotting
bar_width = 0.35
index = np.arange(len(df['Models']))

plt.figure(figsize=(12, 6))
bar1 = plt.bar(index - bar_width/2, df['F1_Score_0'], bar_width, color=colors[0], label='F1 Score 0')
bar2 = plt.bar(index + bar_width/2, df['F1_Score_1'], bar_width, color=colors[1], label='F1 Score 1')

plt.title('F1 Score for Different Models and Classes')
plt.xlabel('Models')
plt.ylabel('F1 Score')
plt.xticks(index, df['Models'], rotation=45, ha='right')
plt.legend()
plt.tight_layout()
plt.show()
```
