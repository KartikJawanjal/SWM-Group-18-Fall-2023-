```python
import pandas as pd
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.feature_extraction.text import CountVectorizer
import nltk
import numpy as np
from helper import remove_stopwords,Word2Vec,Lemmatizer
from sklearn.metrics import classification_report,accuracy_score
from gensim.models import Word2Vec
from xgboost import XGBClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
import pickle
```

```python
df=pd.read_csv("/content/drive/Shareddrives/Semantic Web Mining Stuff/amazon_news_daily_summarized_spacy.csv")
#df=df.dropna()
```

```python
df1=df.dropna()
df1=df1.reset_index(drop=True)
```

```python
from sklearn.model_selection import KFold
vectorizer=CountVectorizer(ngram_range=(2,2), stop_words= ['is','to','my','a','an','the'])
new_dat=vectorizer.fit_transform(df1['text'])
```

```python
with open('Newvectorizer(Amazon Daily).pkl', 'wb') as f:
    pickle.dump(vectorizer, f)
```

```python
kfold=KFold(4,random_state=None,shuffle=True)
#model=LogisticRegression(penalty='l2',class_weight='balanced',solver='lbfgs',verbose=True,max_iter=300)
scores=[]
X=new_dat
y=df1['label']
for train, test in kfold.split(df1):
  X_train,X_test,y_train,y_test=X[train],X[test],y[train],y[test]
  model=LogisticRegression(penalty='l2',class_weight={0: 0.75, 1: 0.25},solver='lbfgs',verbose=True,max_iter=25)
  model.fit(X_train,y_train)
  ypred=model.predict(X_test)
  print(classification_report(y_test,ypred))
  scores.append(model.score(X_test,y_test))
print(scores)
```
