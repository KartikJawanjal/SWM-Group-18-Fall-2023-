```python
import json
count=0
news_list=[]
for zipfolder in list1:
  with zipfile.ZipFile(zipfolder, 'r') as zip1:
    for f_name in zip1.namelist():
        df=pd.DataFrame()
        with zip1.open(f_name) as f:
          count+=1
          data = f.read()
          data=data.decode("utf-8")
          data = json.loads(data)
          if 'apple' in data['title'].lower() or 'amazon' in data['title'].lower() or 'apple' in data['text'].lower() or 'amazon' in data['text'].lower() or 'aapl' in data['title'].lower() or 'amzn' in data['title'].lower() or 'aapl' in data['text'].lower() or 'amzn' in data['text'].lower() :
            new_data = {"title" : data["title"],"text" : data["text"],"published" : data["published"]}
            news_list.append(data)


print(count)
dataframe_prime = pd.DataFrame(news_list, columns=['title','text','published'])
```
